```python
import pandas as pd
import sqlalchemy as sa
```

```python
url = 'mysql+pymysql://root@localhost:3306/canact?charset=utf8'
url_jis = 'mysql+pymysql://root@localhost:3306/canact?charset=utf8'
engine = sa.create_engine(url, echo=False)
engine_jis = sa.create_engine(url_jis, echo=False)
```

```python
query_records = "select * from canact.records where created_at between '2021-01-12 00:00:00' and '2021-02-01 00:00:00' "
records_df = pd.read_sql(query_records,con=engine)

query_activities = "select * from canact.activities where created_at between '2021-01-12 00:00:00' and '2021-02-01 00:00:00' "
activities_df = pd.read_sql(query_activities,con=engine)

query_activity_targets = "select * from canact.activity_targets where created_at between '2021-01-12 00:00:00' and '2021-02-01 00:00:00' "
activity_targets_df = pd.read_sql(query_activity_targets,con=engine)

query_activity_types = "select * from canact.activity_types"
activity_types_df = pd.read_sql(query_activity_types,con=engine)

query_record_types = "select * from canact.record_types"
record_types_df = pd.read_sql(query_record_types,con=engine)
```
